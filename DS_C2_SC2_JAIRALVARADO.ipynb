  a) Extraer información del archivo

```python
from __future__ import print_function
import os
import glob
import numpy as np
import pandas as pd

# Valida si un archivo existe
def ArchivoExiste(archivo):
   return os.path.isfile(archivo)

# Valida si un archivo está vacío
def file_is_not_empty(name):
      return os.stat(name).st_size > 0
#Función que verifica si un archivo existe y no está vacío
def ValidaArchivo(nyrArchivo,solonArchivo):
   ResVal=False
   print ('Verificando si EXISTE archivo de trabajo: ',nyrArchivo)
   if ArchivoExiste(solonArchivo)==True:
      print ('Validando que archivo de trabajo no esté VACíO: ',nyrArchivo)
      if file_is_not_empty(nyrArchivo):
         ResVal=True
      else:
         print('¡Archivo vacío, favor de verificar!')
   else:
      print('¡Archivo NO existe, favor de verificar!')
   return(ResVal)

print('Directorio de trabajo: ', os.getcwd())
ArchivoTrabajo='country_vaccinations.csv'
Nombre_Archivo=os.getcwd()+"/"+ArchivoTrabajo
print('La ruta y nombre del archivo de trabajo deberá ser: ',Nombre_Archivo)
if ValidaArchivo(Nombre_Archivo,ArchivoTrabajo)==True:
   fcv=pd.read_csv('/content/'+ArchivoTrabajo)
   print('Tamaño del resultado',fcv.shape)
   print(fcv.head())
```

b) Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
fcv.info()
fcv.date=pd.to_datetime(fcv.date)
fcv.info()
```

c) Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines).

```python
fcv['VacXcompania']=fcv['vaccines'].str.split(', ').str.len()
print('Cantidad de vacunas de cada compañía:\n', fcv[['vaccines','VacXcompania']])
```

d) Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
print('Total de vacunas aplicadas en todo el mundo: ',fcv['total_vaccinations'].sum())
```

e) Calcular el promedio de vacunas aplicadas por país.

```python
print('Promedio de vacunas aplicadas por país:\n',fcv['total_vaccinations'].groupby(fcv.country).mean())
```

f) Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
vac290121=fcv.loc[fcv['date']=='29-1-2021','total_vaccinations']
print('Vacunas aplicadas a nivel mundial el 29/01/21: ', vac290121.sum())
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccinations y daily_vaccinations_raw.

```python
conDiferencias=fcv
conDiferencias['diferencias']=conDiferencias.daily_vaccinations - conDiferencias.daily_vaccinations_raw
conDiferencias
```

h) Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
print('Fecha de registro mas reciente: ',fcv.date.max())
print('Fecha de registro mas antiguo: ',fcv.date.min())
print('Periodo de tiempo entre ambas fechas: ',fcv.date.max()-fcv.date.min())
```

i) Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /).

```python
conCantidad=fcv
conCantidad['canVac']=conCantidad['vaccines'].str.split('/').str.len()
conCantidad
```

j) Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20=fcv[fcv.date < '20-12-2021']
antes20
```

k) Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer=fcv[fcv['vaccines'].str.contains('Pfizer',case=False)]
pfizer
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
#!pip show xlsxwriter
#!pip show xlwt
#!pip install xlsxwriter
#!pip install xlwt
import pandas as pd
nom_archivo_out='/content/resultadosReto.xlsx'
writer=pd.ExcelWriter(nom_archivo_out,engine="xlsxwriter")
conDiferencias.to_excel(writer,sheet_name='conDiferencias')
conCantidad.to_excel(writer,sheet_name='conCantidad')
antes20.to_excel(writer,sheet_name='antes20')
pfizer.to_excel(writer,sheet_name='Pfizer')
#Guardar cambios en el archivo de salida
writer.close()
print('Dataframe guardados en hojas diferentes en archivo: ',nom_archivo_out)
```
